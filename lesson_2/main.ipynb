## Q&A

Q: Do I need to re-install rails every time I switch my ruby version?

A: Yes, we need to reinstall all our gems every time we switch ruby versions.

Q: Should I experiment with ruby/rails priview versions?

A: At first, probably not: they could completely break an application.

Q: Can we use `rails generate`?

A: At first, we are only allowed to use two generators: `model` and `migration`.
We should only ever use generators if we know exactly what is being generated.
Otherwise we might be introducing bugs and security vulnerabilities without even
realising it.

Q: Should we use Test Driven Deveolpment for everything?

A: Yes, ideally; we should never try to solve the feature all at once. We should
write the test describing the feature and get it to pass incrementally.

Q: Do we need any other gems/dependencies?

A: We won't need any other other gems unless explicitly told otherwise.

#### Implementation

We first generate the model with rails, specifying the name and type of it's attributes:

```bash

```

We use the type `float` to indicate that the attribute should be saved in the 
database as a decimal number as opposed to a string. 

This in turn generates the migration `./db/migrate/20200410014140_create_car_categories.rb`:

````ruby
class CreateCarCategories < ActiveRecord::Migration[6.0]
  def change
    create_table :car_categories do |t|
      t.string :name
      t.float :daily_rate
      t.float :insurance
      t.float :third_party_insurance

      t.timestamps
    end
  end
end
````

That we migrate with `rails db:migrate`.

We then add this resource to `/config/routes.rb` specifying the allowed actions:

````ruby
# frozen_string_literal: true

Rails.application.routes.draw do
  # For details on the DSL available within this file, see https://guides.rubyonrails.org/routing.html
  root 'home#index'

  resources :manufacturers, only: %i[index show new create]
  resources :subsidiaries, only: %i[index show new create]
  resources :car_categories, only: %i[index show new create]
  resources :customers, only: %i[index show new create]
end
````

We follow that up with a controller that corresponds to resource's routes in 

`app/controllers/car_categories_controller.rb`

````ruby
# frozen_string_literal: true

class CarCategoriesController < ApplicationController
  def index
    @car_categories = CarCategory.all
  end

  def show
    @car_category = CarCategory.find(params[:id])
  end

  def new
    @car_category = CarCategory.new
  end

  def create
    @car_category = CarCategory.new(car_category_params)
    if @car_category.save
      flash[:success] = 'Categoria de veículo criada com sucesso'
      redirect_to @car_category
    else
      flash.now[:error] = 'Algo deu errado'
      render :new
    end
  end

  private

  def car_category_params
    params.require(:car_category).permit(:name, :daily_rate, :insurance,
                                         :third_party_insurance)
  end
end
````

The most unusual thing here is the `render` call to the `_form.html.erb`.
As a way of re-utilizing code and better organizing our viewa rails lets us write 
isolated sippets of html that can be called from within other views.

We also see a call to `yield` for `:button_text` in the form, wich takes the 
attribute we declared in `new.html.erb` and directly inserts it in the partial 
without processing/interpreting it first.

Knowing this we can hint as to the difference between `render` and `yield`:
Whatever we call with `render` is processed/interpreted before being inserted in the view.
What we call with `yield` is inserted raw, as is.

### Customers' `:cpf` ans Subsidiaries' `:cnpj` should unique and valid

CPFs and CNPJs are registries that the brazilian government uses to keep track 
of people and companies respectively. 
- CPFs have 11 digits and look like this `679.421.540-33`
- CNPJs have 14 digits and look like this `10.386.453/0001-51`

You can read more about them here:
- https://en.wikipedia.org/wiki/CNPJ
- https://en.wikipedia.org/wiki/Cadastro_de_Pessoas_F%C3%ADsicas

They both have two validator-digits at the end, and with a simple algorithm we
should be able to generate them from previous 9 digits. If the digits we generate
match the validat-digits we know that the cpf/cnpj is valid. We we'llget into the
details of this algorithm once we express it in thr code.

We start, as always by writing a test for cnpj in `spec/features/subsidiaries_cnpj_spec.rb`:

#### Implementation

The first thing we do is create two migrations to index the `cnpj` and `:cpf` 
columns in their respective tables. We need to do this for two reasons:
1. It communicates the uniqueness of the record to the database, which is 
external to our rails application;
2. Whenever we create a record we need to hit the database and verify whether it 
has been taken. If these two columns aren't indexed the database would have to 
do a full table scan in order to verify this, which is extremely heavy on performance.

We can use rails generate as follows:

```bash
$ bundle exec rails generate migration add_index_to_subsidiary_cnpj

```
and fill them respectively:

````ruby
class AddIndexToSubsidiaryCnpj < ActiveRecord::Migration[6.0]
  def change
    add_index :subsidiaries, :cnpj, unique: true
  end
end
````

The only odd ones here are `cnpj: true` and `cpf: true`: why would it be validating itself?

We proceed to write the validators, `app/validators/cnpj_validator.rb` and `app/validators/cpf_validator.rb`:

````ruby
# frozen_string_literal: true

# Validates the format of "Cadastro Nacional da Pessoa Juridica", a Brazilian
# taxpayer registry identification for businesses and companies, issued by the
# Brazilian Department of Federal Revenue.
class CnpjValidator < ActiveModel::EachValidator
  BLACKLIST = %w[
    00000000000000
    11111111111111
    22222222222222
    33333333333333
    44444444444444
    55555555555555
    66666666666666
    77777777777777
    88888888888888
    99999999999999
  ].freeze

  def validate_each(record, attribute, value)
    return fail_validation(record, attribute) if value.blank?
    return fail_validation(record, attribute) if in_blacklist?(value)

    raw_cnpj = parse_cnpj(value)

    fail_validation(record, attribute) unless valid_digits?(raw_cnpj)
  end

  private

  def fail_validation(record, attribute)
    record.errors[attribute] << (options[:message] || ' não é valido')
  end

  def in_blacklist?(value)
    BLACKLIST.include?(value)
  end

  def parse_cnpj(value)
    value.each_char.to_a.map(&:to_i)
  end

  def valid_digits?(raw_cnpj)
    digits = raw_cnpj[0...12]
    digits << generate_verification_digit(digits)
    digits << generate_verification_digit(digits)

    digits[-2, 2] == raw_cnpj[-2, 2]
  end

  def generate_verification_digit(digits)
    index = 2

    sum = digits.reverse.reduce(0) do |buffer, number|
      (buffer + number * index).tap do
        index = index == 9 ? 2 : index + 1
      end
    end

    mod = sum % 11
    mod < 2 ? 0 : 11 - mod
  end
end
````

````ruby
# frozen_string_literal: true

# Validates the format of "Cadastro de Pessoas Fisicas", a Brazilian individual
# taxpayer registry identification for citizens and resident aliens, issued by
# the Brazilian Department of Federal Revenue.
class CpfValidator < ActiveModel::EachValidator
  BLACKLIST = %w[
    00000000000
    11111111111
    22222222222
    33333333333
    44444444444
    55555555555
    66666666666
    77777777777
    88888888888
    99999999999
    12345678909
    01234567890
  ].freeze

  def validate_each(record, attribute, value)
    return fail_validation(record, attribute) if value.blank?
    return fail_validation(record, attribute) if in_blacklist?(value)

    raw_cpf = parse_cpf(value)

    fail_validation(record, attribute) unless valid_digits?(raw_cpf)
  end

  private

  def fail_validation(record, attribute)
    record.errors[attribute] << (options[:message] || ' não é valido')
  end

  def in_blacklist?(value)
    BLACKLIST.include?(value)
  end

  def parse_cpf(value)
    value.each_char.to_a.map(&:to_i)
  end

  def valid_digits?(raw_cpf)
    digits = raw_cpf[0...9]
    digits << generate_verification_digit(digits)
    digits << generate_verification_digit(digits)

    digits[-2, 2] == raw_cpf[-2, 2]
  end

  def generate_verification_digit(digits)
    modulus = digits.size + 1

    multiplied = digits.map.each_with_index do |number, index|
      number * (modulus - index)
    end

    mod = multiplied.reduce(:+) % 11
    mod < 2 ? 0 : 11 - mod
  end
end
````

Now there's a lot going on here, and we'll only analyze the cpf validator for simplicity
and also because they're both pretty similar. At first glance, the class inherits from 
`ActiveModel::EachValidator`, which is the easiest way to do custom validations in rails
(there are others). 

We then see a `BLACKLIST` constant, an array that contains cpfs that would 
otherwise pass our validation but are deemed not valid for being too generic. 
No brazilian resident has or should ever have a cpf in this array.
Outside the class this constant can be accessed as `CpfValidator::BLACKLIST`.

````ruby
def generate_verification_digit(digits)
  modulus = digits.size + 1

  multiplied = digits.map.each_with_index do |number, index|
    number * (modulus - index)
  end

  mod = multiplied.reduce(:+) % 11
  mod < 2 ? 0 : 11 - mod
end

raw_cpf = "12345678909".each_char.to_a.map(&:to_i)

digits = raw_cpf[0...9]
digits << generate_verification_digit(digits)
digits << generate_verification_digit(digits)

puts "Is #{raw_cpf.join} valid? #{digits[-2, 2] == raw_cpf[-2, 2]}"
puts "It shouldn't be!"
````

The class and its methods can then be conceptually divide in two categories: 

Methods that interact with the rails API
  - `validate_each`: orchestrates the flow of the validation, calling the verifications
  and invalidating `value` as necessary. One thing to note is the `return` call on a
  `value.blank?`. We call it at the beggining of the method so to stop the other 
  validations, which which would error out otherwise due to `nil` values. We also
  fail the validation with `in_blacklist?(value)` for obvious reasons.

  - `fail_validation`: takes in an `attribute` hash (where active model keeps 
  track of the model's validations) and adds to it the error message. The `record`
  key references the attribute that failed. We could add here any failure message 
  we want, and later retrieve it in the view.

"Low-level" methods that perform math and string operations.
  - `parse_cpf`: takes the `value` string and returns it as an array of integers. We
  do this so we can later call math operations on these numbers.

````ruby
def parse_cpf(value)
  value.each_char.to_a.map(&:to_i)
end

parsed_cpf = parse_cpf("78108243254")

puts parsed_cpf 
````

- `in_blacklist?`: Checks whether the cpf is blacklisted, returning a boolean.
(By the way, we can re-assign a constant in ruby)

````ruby
  BLACKLIST = %w[
    00000000000
    11111111111
    22222222222
    33333333333
    44444444444
    55555555555
    66666666666
    77777777777
    88888888888
    99999999999
    12345678909
    01234567890
  ].freeze

def in_blacklist?(value)
  BLACKLIST.include?(value)
end

in_blacklist_1 = "55555555555"
not_in_blacklist_1 = "78108243254"
in_blacklist_2 = "12345678909"
not_in_blacklist_2 = "78118213454"

puts in_blacklist?(in_blacklist_1)
puts in_blacklist?(not_in_blacklist_1)
puts in_blacklist?(in_blacklist_2)
puts in_blacklist?(not_in_blacklist_2)
````

  - `valid_digits?`: take for example the valid cpf 384.851.908-90; the last two 
  digits "9" and "0" are separated with a dash because they're special: they're the validation
  digits. Cpfs, being official documents and all, are designed in such a way so as 
  to easily tell if even one of the digits was mistaken during transcription.
  
  In a nutshell, this method goes through the algorithm that generates the verification digits
  from from the first 9 digits, "384851908", and compares them with the ones provided, "90".
  If they're the same, we know this is a valid cpf and it returns true.
 
  - `generate_verification_digit`: generates a single verification digit from a cpf integer array.
  To do this it first takes the sum of the first nine digits of the cpf multiplied by a decreasing 
  natural number sequence from ten to 2, like so:

````ruby
first_validation_sum =  3 * 10 +  8 * 9  +  4 * 8 +  8 * 7 +  5 * 6 +  1 * 5 +  9 * 4 +  0 * 3 +  8 * 2
puts first_validation_sum
````

  we then multiply by ten and divide by 11. The remainder/modulos of this operation 
  equals the first validation digit of our cpf if it's valid, 9 in this case.

````ruby
first_validation_digit = first_validation_sum * 10 % 11
puts first_validation_digit
````

  To get the second digit we do the same thing but with the sequence 11 to 2 and 
  with the first validation digit.

````ruby
second_validation_sum = 3 * 11 +  8 * 10 +  4 * 9 +  8 * 8 +  5 * 7 +  1 * 6 +  9 * 5 +  0 * 4 +  8 * 3  + first_validation_digit * 2
puts second_validation_sum

second_validation_digit = second_validation_sum * 10 % 11
puts second_validation_digit
````

  We see that the first and second validation digits we generated match the ones
  in the cpf, and we can now be sure it's valid.

````ruby
puts first_validation_digit.to_s + second_validation_digit.to_s == '90'
````

These two functions achieve all this in a few stubby lines thanks to ruby's rich
syntax and utility methods:
  - with `[0...9]` we select the first nine elements from the array, with indexes
  0, 1, 2, 3, 4, 5, 6, 7 and 8 (arrays are zero-indexed). The three point `...`
  tell ruby to iterate *up to* eight, or to the second-before-last. If we want to 
  iterate up to the last number we use `[0..9]`.
  - the shovel operator `<<` shoves elements at the end of an array
  - `[-2, 2]` takes the last two elements of an array. The index of the last element of an 
  array is always `-1`, and the index of the second-before-last `-2`. We're basically telling ruby 
  to go to the second-before-last element and take the next `2` elements including
  the second-before-last.

````ruby
  def valid_digits?(raw_cpf)
    digits = raw_cpf[0...9]
    digits << generate_verification_digit(digits)
    digits << generate_verification_digit(digits)

    digits[-2, 2] == raw_cpf[-2, 2]
  end

up_to_before_last_iteration = (0...9)
up_to_last_iteration = (0..9)

puts up_to_before_last_iteration
puts up_to_last_iteration

up_to_before_last_array = up_to_before_last_iteration.to_a
up_to_last_array = up_to_last_iteration.to_a

puts up_to_before_last_array
puts up_to_last_array

up_to_before_last_array << 'I like'
up_to_last_array << 'turtles'

puts up_to_before_last_array
puts up_to_last_array

last_couple_elements_of_up_to_before_last_array = up_to_before_last_array[-2, 2]
last_couple_elements_of_up_to_last_array = up_to_last_array[-2, 2]

puts last_couple_elements_of_up_to_before_last_array
puts last_couple_elements_of_up_to_last_array
````

  - `.size` returns the size of the array, the number of elements.
  - `.map` returns an array out of return of each iteration in its block
  - `.each_with_index` exposing the element's index and value in the block
  - `.reduce(:+)` resturns the sum of all the elements of an array. We can pass
  it any other `:symbol` method, like `:-` for the diference of all the elements,
  `:*` for the product or `:/` for the quotient. Very handy method indeed.
  - the modulos operator `%` returns the integer remainder of the quotient.
  - we finally have the ternary operator `mod < 2 ? 0 : 11 - mod`, that evaluates 
  whether the first condition `mod < 2` is true: 
  
    - *if it is true it'll return `0`*
    - *if it's false it'll return `11 - mod`*
   
  Fundamentally, it's a compact way of writing an `if ... else ... end` block.

````ruby
  def generate_verification_digit(digits)
    modulus = digits.size + 1

    multiplied = digits.map.each_with_index do |number, index|
      number * (modulus - index)
    end

    mod = multiplied.reduce(:+) % 11
    mod < 2 ? 0 : 11 - mod
  end

sample_digits = (0...9).to_a
puts sample_digits
puts sample_digits.size

sample_modulus = sample_digits.size + 1
puts sample_modulus

sample_multiplied = sample_digits.map.each_with_index do |number, index|
  number * (sample_modulus - index)
end
puts sample_multiplied

sample_mod = sample_multiplied.reduce(:+) % 11
puts sample_multiplied

sample_return = sample_mod < 2 ? 0 : 11 - sample_mod
puts sample_return
````
